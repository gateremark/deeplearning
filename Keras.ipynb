### Keras Sequential Model

This is a linear stack of layers.

```python
from keras.models import Sequential
from keras.layers import Dense, Activation

#Dense is a type of layer (The most basic type of layer in a neural network)
model = Sequential ([
                     Dense(32, input_shape=(10,), activation='relu'),
                     Dense(2, activation='softmax'),
])

# Dense(number on nodes/neurons, shape of data being passed into the model, non-linear function)
```
